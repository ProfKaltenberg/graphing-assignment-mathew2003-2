```python
#Question1: Creating 2 Graphs 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
path = r'C:\Users\mathe\OneDrive\Documents\GitHub\assignment-1-mathew2003'
data = pd.read_csv(path + r'\hud_fmr_vacancy_income_county.csv')
print(data.head())
print(data.columns.tolist())
```

```python
#Question2, 3, and 4:
fig, ax = plt.subplots(nrows=1, ncols=2)

ax1 = plt.subplot(1, 2, 1)
ax1.plot(data['vacancy_rate'], data['fmr_2bed'], linestyle='none', marker='.', color='cornflowerblue')
ax1.set_xlabel('Amount of Vacant Apartments')
ax1.set_ylabel('Rent for 2 Bedroom Apartment ($)')
ax1.set_title('How Vacancy Rate Affects 2-Bedroom Rent         ')

# --- Graph 2: Histogram (one variable at a time) ---
ax2 = plt.subplot(1, 2, 2)
ax2.hist(data['median_income'], bins=50)
ax2.set_xlabel('median income')
ax2.set_ylabel('Number of People')
ax2.set_title('Various Incomes of People')

fig.suptitle('ScatterPlot and Histograph', fontsize=16)
fig.tight_layout()
plt.show()

#The graph on the left is a scatterplot and it shows a slight positive correlation between Rent and Vacancy rate. 
#On the right there is a histograph and it shows that most people in the dataset make around 60k a year. 
```

```python
#Question 5:
#Pie Chart: This chart shows the top 8 states with apartments that are actually for rent
#**note** Vacancy rate is the number of apartment that are actually available for rent, and total vacant is the total amount of apartment that are vacant and includes apartments that are not for rent.
vacant_by_state = data[['state', 'vacancy_rate']].groupby('state').sum().reset_index()
top_states = vacant_by_state.nlargest(8, 'vacancy_rate')

fig, ax1 = plt.subplots()
ax1.pie(top_states['vacancy_rate'], labels=top_states['state'], autopct='%1.1f%%', startangle=90)
ax1.set_title('Top 8 States with apartments for rent')
plt.show()
```
